### 1. Abre este script y explica su contenido. Podrás ver que se utiliza el primer dataset de ejemplo y se realiza la representación gráfica del SVM. Comenta que tipo de kernel se está utilizando y cuáles son los parámetros de entrenamient

Una vez abrimos el script, podemos observar que lo primero que hay que hacer es importar las librerías necesarias para poder realizar nuestra práctica.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn import svm
```

Una vez importadas, tendremos que leer los datos de nuestra base de datos y guardarlos en unas variables. Primero, guardaremos todos los valores de entrada(X) menos la clase a la perteneciente, la cuál la guardaremos en otra variable(y)

```python
data = pd.read_csv("BasesDatos/csv/dataset1.csv", header=None)

X = data.iloc[:,:-1].values
y = data.iloc[:,-1].values
```

Seguidamente, procedemos a crear nuestro modelo SVM, el cuál se creará con la función svm.SVC. Con respecto al kernel que utilizaremos, será un kernel de tipo RBF, el cuál es el más usado en cuanto a la creación de SVMs. Esta función necesitará, además del tipo de kernel, dos parámetros. 

El primero será C, el cuál se encargará de indicarnos el error cometido por nuestra red, un mayor valor de C indica que vamos a conseguir el menor número de errores. En cambio, a un menor valor indica que vamos a conseguir un mayor margen pero cometiendo errores. 

El segundo valor será gamma, el cuál es el coeficiente de nuestro kernel. Este parámetro indica la suavidad con la que obtendremos la solución, a mayor valor de gamma, menor es la suavidad con la que obtendremos esa solución, además de que esto nos conducirá a un modelo con un alto sesgo y una baja varianza. A valores bajos ocurre lo contrario, es decir, un modelo con bajo sesgo pero alta varianza.

Finalmente, entrenaremos el modelo con la función 'fit', la cuál se entrenará con los valores anteriormente guardados (X, y)

```python
svm_model = svm.SVC(kernel='rbf',C=200,gamma=200)
svm_model.fit(X, y)
```

Por último, y para finalizar la descripción del script original, mostraremos los resultados, calculando cuáles son los límites de nuestro modelo y además, dibujando cada uno de los límites sobre todos los patrones de nuestro conjunto de datos.

```python
# Representar los puntos
plt.figure(1)
plt.clf()
plt.scatter(X[:, 0], X[:, 1], c=y, zorder=100, cmap=plt.cm.Paired, edgecolors='black')

# Representar el hiperplano separador
plt.axis('tight')
# Extraer límites
x_min = X[:, 0].min()
x_max = X[:, 0].max()
y_min = X[:, 1].min()
y_max = X[:, 1].max()

# Crear un grid con todos los puntos y obtener el valor Z devuelto por la SVM
XX, YY = np.mgrid[x_min:x_max:500j, y_min:y_max:500j]
Z = svm_model.decision_function(np.c_[XX.ravel(), YY.ravel()])

# Hacer un plot a color con los resultados
Z = Z.reshape(XX.shape)
plt.pcolormesh(XX, YY, Z > 0, cmap=plt.cm.Paired)
plt.contour(XX, YY, Z, colors=['k', 'k', 'k'], linestyles=['--', '-', '--'],
                levels=[-.5, 0, .5])

plt.show()
```

### 6. Propón una configuración de SVM no lineal (utilizando el kernel tipo RBF o Gaussiano) que resuelva el problema. El resultado debería ser similar al de la Figura 3. ¿Qué valores has considerado para C y para γ?. Además, incluye un ejemplo de una configuración de parámetros que produzca sobre-entrenamiento y otra que produzca infra-entrenamiento.

Para realizar este ejercicio, vamos a utilizar un kernel de tipo RBF, e iremos probando hasta conseguir que los patrones se clasifiquen corresctamente

```python
svm_model = svm.SVC(kernel='rbf', C=100, gamma=140)
svm_model.fit(X, y)
# Representar los puntos
plt.figure(1)
plt.clf()
plt.scatter(X[:, 0], X[:, 1], c=y, zorder=100, cmap=plt.cm.Paired, edgecolors='black')

# Representar el hiperplano separador
plt.axis('tight')
# Extraer límites
x_min = X[:, 0].min()
x_max = X[:, 0].max()
y_min = X[:, 1].min()
y_max = X[:, 1].max()

# Crear un grid con todos los puntos y obtener el valor Z devuelto por la SVM
XX, YY = np.mgrid[x_min:x_max:500j, y_min:y_max:500j]
Z = svm_model.decision_function(np.c_[XX.ravel(), YY.ravel()])

# Hacer un plot a color con los resultados
Z = Z.reshape(XX.shape)
plt.pcolormesh(XX, YY, Z > 0, cmap=plt.cm.Paired)
plt.contour(XX, YY, Z, colors=['k', 'k', 'k'], linestyles=['--', '-', '--'],
                levels=[-.5, 0, .5])

plt.show()
print("The score is:", svm_model.score(X, y , sample_weight=None)*100, "%")
```

Como vemos, con C=100 y gamma=140 conseguimos un 100% de patrones bien clasificados, siendo esto lo que estábamos buscando con una SVM con un kernel lineal. Estos valores no son ni muy altos ni muy bajos, con lo que esta configuración sería la ideal para este conjunto de datos.

##### 1. Sobre-entrenamiento

```python
svm_model = svm.SVC(kernel='rbf', C=200, gamma=1200)
svm_model.fit(X, y)
# Representar los puntos
plt.figure(1)
plt.clf()
plt.scatter(X[:, 0], X[:, 1], c=y, zorder=100, cmap=plt.cm.Paired, edgecolors='black')

# Representar el hiperplano separador
plt.axis('tight')
# Extraer límites
x_min = X[:, 0].min()
x_max = X[:, 0].max()
y_min = X[:, 1].min()
y_max = X[:, 1].max()

# Crear un grid con todos los puntos y obtener el valor Z devuelto por la SVM
XX, YY = np.mgrid[x_min:x_max:500j, y_min:y_max:500j]
Z = svm_model.decision_function(np.c_[XX.ravel(), YY.ravel()])

# Hacer un plot a color con los resultados
Z = Z.reshape(XX.shape)
plt.pcolormesh(XX, YY, Z > 0, cmap=plt.cm.cool)
plt.contour(XX, YY, Z, colors=['k', 'k', 'k'], linestyles=['--', '-', '--'],
                levels=[-.5, 0, .5])

plt.show()
print("The score is:", svm_model.score(X, y , sample_weight=None)*100, "%")
```

Para apreciar mejor el sobre-entrenamiento, hemos cambiado el color de la gráfica. Como vemos, a mayor valor de gamma(1200), se obtiene un sobre-entrenamiento, esto es debido a que la "suavidad" con la que se obtiene la solución es casi nula.

### 12. Para estar seguros de que has entendido cómo se realiza la búsqueda de parámetros, implementa de forma manual la validación cruzada anidada tipo K-fold expuesta en esta sección. Te puede ser útil la función np.meshgrid, el uso de listas por compresión y la clase StratifiedKFold. Compara los resultados con los que obtienes usando GridSearchCV.

Para este ejercicio vamos a utilizar una nueva función de sklearn llamada StratifiedKFold, la cual dividirá nuestro conjunto de datos en K conjuntos para realizar la validación cruzada. Lo primero que hay que hacer es indicar el número de conjuntos que queremos para la validación.

```python
K = 15
```

Seguidamente, crearemos un conjunto de valores para que no sea todo manual. Estos valores serán los correspondientes para C y gamma.

```python
C_values = np.logspace(-5, 15, num=11, base=2)
G_values = np.logspace(-15, 3, num=9, base=2)
```

Una vez hecho esto, realizaremos la partición de nuestro conjunto, en K partes, con la función StratifiedKFold.

```python
from sklearn.model_selection import StratifiedKFold
skf = StratifiedKFold(n_splits=K)
```

Realizada la partición del conjunto, procederemos al entrenamiento con validación. Para ello, guardaremos los mejores resultados de entre todas las combinaciones y lo mostraremos por pantalla.

```python
best_CCR = 0
gamma = 0
C = 0

for c in C_values:
    for g in G_values:
        scores = []
        #Creamos el modelo
        svm_model = svm.SVC(kernel='rbf', C=c, gamma=g)
        
        for train_index, test_index in skf.split(X, y):
            X_test = X[test_index]
            X_train = X[train_index]
            y_test = y[test_index]
            y_train = y[train_index]
            
            #Entrenamos el modelo
            svm_model.fit(X_train, y_train)
            scores.append(svm_model.score(X_test, y_test))
            
        scores = np.array(scores)
        #Calculamos la media de los CCR obtenidos
        score = np.mean(scores)
        if score > best_CCR:
            best_CCR = score
            C = c
            gamma = g
        
print("El mejor modelo es: C=", C, "gamma=", gamma, "score=", best_CCR*100,"%")
```
